### MAP, FILTER 
- map(fonksiyon,liste,...)
- map() fonksiyonu parametre olarak aldığı fonksiyona , parametre olarak aldığı listenin her elemanını sırasıyla parametre olarak gönderir.
filter() fonksiyonu ise çağırılan fonksiyonun döndürdüğü değerin true olduğu durumlara göre liste döndürür. Yani fonksiyona gönderilen değerler içerisinden istediklerimizi alarak yeni liste oluşturmamızı sağlar. 

```python
def kare(sayi):
    return sayi*sayi
sayilar = [1,2,3,4,5]
sonuc = map(kare,sayilar)
print(list(sonuc))
```

```python
def teksayi(sayi):
    return sayi%2==1
    
liste = range(1,20)
print(list(filter(teksayi,liste)))
```

```python
lst=[1,2,3,4]
# map (fonsksiyoni liste)
list(map((lambda x: x**3), lst))
```

```python
# filter(fonksiyon, liste)
list(filter((lambda x: x**3), lst))
```
