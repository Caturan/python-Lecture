```python
def histogram(deger):
    hist_dict = {}
    for k in deger: 
        if k in hist_dict:
            hist_dict[k] += 1 
        else: 
            hist_dict[k] = 1 
    return hist_dict 

print(histogram("Cevdet"))
```

```python
numbers = range(11)
new_dict_for = {}

for n in numbers:
    if n%2==0:
        new_dict_for[n] = n**2

print(new_dict_for)
```

```python
new_dict_comp = {n:n**2 for n in numbers if n%2 == 0}

print(new_dict_comp)
```

```python
dict1 = {'a': 1, 'b': 2, 'c': 3, 'd': 4, 'e': 5}

dict1_cond = {k:v for (k,v) in dict1.items() if v>2}

print(dict1_cond)
```

```python
dic1_doubleCond = {k:v for (k,v) in dict1.items() if v>2 if v%2 == 0}

print(dic1_doubleCond)
```

```python
dict1 = {'a': 1, 'b': 2, 'c': 3, 'd': 4, 'e': 5, 'f':6}

dict1_tripleCond = {k:('çift' if v%2==0 else 'tek') for (k,v) in dict1.items()}

print(dict1_tripleCond)
```
